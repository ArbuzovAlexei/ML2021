```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('vgsales.csv', sep=",")
data.head()
```

Рассмотрим распределение игр по годам выпусков:

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.kdeplot(data=data, x="Year")
```

Больше всего игр было выпущено в 2008 - 2010 годах

Расмотрим число игр различных жанров:

```python
fig, ax = plt.subplots(figsize=(20,10)) 
sns.histplot(data['Genre'], discrete=True)
```

Больше всего игр "Экшенов"

Рассмотрим распределение жанров игр по годам

```python
fig, ax = plt.subplots(figsize=(20,10)) 
sns.histplot(data, x="Genre", y="Year", discrete=True)
```

Аналогично диаграмма-скрипка

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.violinplot(x='Year', y='Genre', data=data)
```

Стратегии появились только в 90-е, до этого данных нет.
<br>Первая стратегия вышла в 1995 (90-х) и с нее начался этот жанр игр

Корреляционная матрица

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(data.corr(method='pearson'), annot=True, fmt='.2f')
```
